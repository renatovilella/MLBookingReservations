```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('..\\NewHotel.csv')

# Replace 'Canceled' and 'Not_Canceled' with an empty string ('') throughout the DataFrame
df = df.replace({'Canceled': '', 'Not_Canceled': '', 'Not_': ''}, regex=True)

# Save the modified DataFrame to a new CSV file
df.to_csv('ModifiedReservation.csv', index=False)
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
data = TabularDataset("..\\Hotel Reservations.csv")
datatocomplete = TabularDataset("..\\ModifiedReservation.csv")
```

```python
data.head()
```

```python
datatocomplete.head()
```

```python
len(data)
```

```python
train_size = 30000
seed = 42
```

```python
train_data = data.sample(train_size, random_state=seed)
test_data = data.drop(train_data.index)
```

```python
train_data.head()
len(train_data)
```

```python
test_data.head()
len(test_data)
```

```python
save_path = 'reservation_predictors'
```

```python
predictor = TabularPredictor(label="booking_status", path=save_path)
```

```python
predictor.fit(train_data)
```

```python
predictor.fit_summary()
```

```python
y_test = test_data["booking_status"]
test_data = test_data.drop(columns=["booking_status"])
```

```python
y_pred = predictor.predict(test_data)
```

```python
metrics = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```
